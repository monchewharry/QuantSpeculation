## peak detection

```python
import numpy as np
from peakutils.peak import indexes
#http://pythonhosted.org/PeakUtils/reference.html#peakutils.peak.indexes
#Finds the peaks by taking its first order difference.

%matplotlib inline
import matplotlib.pyplot as plt
import tushare as tu
```

```python
vector = tu.get_hist_data('600606',start='2016-01-23',end='2016-12-23')['close']
vector.plot(figsize=(12,8)); #method
plt.title('600606 close price') #price chart
```

```python
loc = indexes(vector.values, thres=4.0/max(vector), min_dist=5)
print('Peaks are: %s' % (loc))
plt.figure(figsize=(12,8))
plt.plot(vector,'^',ls='-',markevery=loc)
```

```python
loc = indexes(vector.values, thres=2.0/max(vector), min_dist=5)
loc2 = indexes(-vector.values, thres=2.0/max(vector), min_dist=5)

print('Peaks are: %s' % (loc))
print('bottoms are: %s' % (loc))
plt.figure(figsize=(12,8))
plt.plot(vector,'^',ls='-',markevery=loc)
```

```python
loc_m=np.concatenate((loc,loc2))
sym_m = ['^'] * len(loc)+ ['s'] * len(loc2)

fig, ax = plt.subplots()
fig.set_size_inches(12, 10)
ax.set_xlim(0,len(vector))
ax.plot(vector)
for l,s in zip(loc_m,sym_m):
    ax.scatter(l,vector.iloc[l],marker=s)
#ax.set_xticklabels()
```
